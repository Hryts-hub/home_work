```python
from kivy.app import App
from kivy.uix.button import Button
from kivy.uix.boxlayout import BoxLayout
from kivy.uix.label import Label


class MyApp(App):
    
    
    def build(self):
        
        widget = BoxLayout(orientation="vertical", padding=30) 
        
        self.label = Label(font_size=55)
        self.bufer_label = Label(font_size=50)
        
        widget.add_widget(self.bufer_label)
        widget.add_widget(self.label)
        
        arr = [
            ['(', ')', '<-', 'C'],
            ['7', '8', '9', '+'],
            ['4', '5', '6', '-'],
            ['1', '2', '3', '*'],
            ['=', '0', '.', '/'],
                ]

        for row in arr:
            row_widget = BoxLayout(orientation="horizontal")
            for symbol in row:
                btn = Button(text=symbol, on_press=self.click, font_size=55)
                row_widget.add_widget(btn)
            widget.add_widget(row_widget)
            
        return widget


    def click(self, instance):
        
        if self.bufer_label.text:
            if self.bufer_label.text.startswith('*') or self.bufer_label.text.startswith('/'):
                self.bufer_label.text = self.bufer_label.text[1:]
        
        if self.label.text:
            if self.label.text.startswith('.'):
                self.label.text = "0" + self.label.text
            if self.label.text == '=error':
                self.label.text = ""
            elif self.label.text[0] == '=':
                self.bufer_label.text = ""
                self.label.text = self.label.text[1:]
        
        if instance.text == '<-':
            if self.label.text:
                self.label.text = self.label.text[:-1]
            else:
                self.bufer_label.text = self.bufer_label.text[:-1]
            
        if instance.text.isdigit() or instance.text in {'(', ')', '.'}:
            self.label.text += instance.text         
        
        if instance.text in {'+', '-', '*', '/'}: 
            if (self.bufer_label.text.endswith("*") or
                self.bufer_label.text.endswith("/") or
                self.bufer_label.text.endswith("-") or
                self.bufer_label.text.endswith("+")
               ) and not self.label.text:
                self.bufer_label.text = self.bufer_label.text[:-1] + instance.text
                bufer_string = ""
            else:         
                bufer_string = self.label.text + instance.text
                
            if bufer_string.count(".") > 1:
                i = bufer_string.index(".")
                bufer_string = bufer_string.replace(".", "")
                point_str = "."
                bufer_string = f"{bufer_string[:i]}{point_str}{bufer_string[i:]}" 
                
            self.bufer_label.text += bufer_string          
            self.label.text = ""            
            
        if instance.text == '=':
            self.bufer_label.text += self.label.text
            
            if (self.bufer_label.text.endswith("*") or
                self.bufer_label.text.endswith("/") or
                self.bufer_label.text.endswith("-") or
                self.bufer_label.text.endswith("+")
               ):
                self.bufer_label.text = self.bufer_label.text[:-1]
            try:
                result = str(eval(self.bufer_label.text))
            except Exception:
                result = "error"
            self.label.text = '=' + result            
              
        if instance.text == 'C':
            self.label.text = ""
            self.bufer_label.text = ""                
                

MyApp().run()
```
